```python
# 라이브러리 불러오기
import pandas as pd
import folium


#  데이터프레임 변환
df = pd.read_csv('/content/도로교통공단_사망 교통사고 정보_20211231.csv',  encoding = 'CP949')  
df=df[df['발생지시도'] == '인천'] # 인천광역시 교통사고
print(df)

# 지도 만들기
smap = folium.Map(location=[ 37.58518054263194 , 126.59432251898228  ], tiles='Stamen Terrain',zoom_start=11)
for name, lat, lng ,cnt in zip(df.발생년월일시, df.위도, df.경도 ,df.사망자수):
    folium.Circle([lat, lng], popup=name, radius=cnt*200, color='red',fill_color = 'red').add_to(smap)
    
smap
```
